```python
# %% [markdown]
# # Legal Document Summarizer - Experimentation Notebook
# 
# This notebook demonstrates how to use the legal summarizer system.

# %% [markdown]
# ## 1. Setup and Imports

# %%
import sys
import os
sys.path.append('..')

from src.models.bart_summarizer import BartSummarizer
from src.models.pegasus_summarizer import PegasusSummarizer
from src.data_processing.dataset_loader import load_and_prepare_data
from src.inference.generate_summary import SummaryGenerator
import pandas as pd

# %% [markdown]
# ## 2. Test with Sample Legal Text

# %%
# Sample legal text
sample_text = """
The plaintiff, John Doe, hereby files this complaint against the defendant, XYZ Corporation, 
for breach of contract and negligence. The plaintiff alleges that on or about January 15, 2023, 
the defendant entered into a written agreement with the plaintiff to provide consulting services. 
The defendant failed to perform their obligations under said agreement, causing damages in the 
amount of $500,000. The plaintiff seeks compensatory damages, legal fees, and any other relief 
the court deems proper.
"""

# %% [markdown]
# ### Try BART Summarizer

# %%
# Initialize BART summarizer
bart_summarizer = BartSummarizer()
summary = bart_summarizer.summarize(sample_text)
print("BART Summary:")
print(summary)

# %% [markdown]
# ### Try PEGASUS Summarizer

# %%
# Initialize PEGASUS summarizer
pegasus_summarizer = PegasusSummarizer()
summary = pegasus_summarizer.summarize(sample_text)
print("PEGASUS Summary:")
print(summary)

# %% [markdown]
# ## 3. Load and Explore Dataset

# %%
# Load CNN/DailyMail dataset
train_df, val_df, test_df = load_and_prepare_data()

# Display sample
print("Sample article:")
print(train_df['article'].iloc[0][:500] + "...")
print("\nSample summary:")
print(train_df['highlights'].iloc[0])

# %% [markdown]
# ## 4. Batch Processing

# %%
# Test batch summarization
test_texts = [
    "The court hereby orders that the defendant shall pay restitution in the amount of $50,000 to the plaintiff within 30 days of this judgment.",
    "Pursuant to Rule 12(b)(6) of the Federal Rules of Civil Procedure, the defendant moves to dismiss the complaint for failure to state a claim upon which relief can be granted."
]

generator = SummaryGenerator(model_type='bart')
summaries = generator.summarizer.summarize_batch(test_texts)

for i, (text, summary) in enumerate(zip(test_texts, summaries)):
    print(f"\nText {i+1}: {text}")
    print(f"Summary {i+1}: {summary}")

# %% [markdown]
# ## 5. Compare Models

# %%
def compare_models(text):
    """Compare BART and PEGASUS summaries"""
    bart = BartSummarizer()
    pegasus = PegasusSummarizer()
    
    print("Original Text:")
    print(text[:300] + "...\n")
    
    print("BART Summary:")
    print(bart.summarize(text))
    print("\nPEGASUS Summary:")
    print(pegasus.summarize(text))

# Test comparison
compare_models(sample_text)
```
